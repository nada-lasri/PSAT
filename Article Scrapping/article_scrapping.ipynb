```python
#A partir du scrapper de l'année dernière, extraction du titre+text:
from WebScrapperWithPreprocess import ArticleScrapperWithDownload

def scrapper19(url):
    try:
        title = ""
        text = ""
        scrapper = ArticleScrapperWithDownload(url)
        scrapper.preprocessAndExtraction()

        if title is None or text is None:
            continue
        
        title = scrapper.article.title
        text = scrapper.article_text
        
        return title,text
    
    except Exception as e:
        print(e)
```

```python
#Extraction pour chaque lien du csv des attributs title et texte que nous renvoie l'algo de 2019
comparaison_df = df.drop(['id','created_at', 'updated_at', 'deleted_at'], axis=1)
comparaison_df_small = comparaison_df.head(20)
# comparaison_df_small = comparaison_df
results_title = []
results_text = []

for index, row in comparaison_df_small.iterrows():
    title, text = "",""
    url = row['url']
#     print(url)
    try:
        title, text = scrapper19(url)
#     print(title)
#     print(text)
        
    except Exception as e:
        continue
#     print(title)
#     print(text)
    results_title.append(title)
    results_text.append(text)
    
comparaison_df_small
```

```python
comparaison_df_small["scrapped_title"] = results_title
comparaison_df_small["scrapped_text"] = results_text
```

```python
#Affichage du dataframe contenant le title et le link de référence, ainsi que le title et le link scrapés
comparaison_df_small
```

```python
#implementation du calcul de la distance cosinus entre un texte 1 et un texte 2
import nltk, string
from sklearn.feature_extraction.text import TfidfVectorizer

stemmer = nltk.stem.porter.PorterStemmer()
remove_punctuation_map = dict((ord(char), None) for char in string.punctuation)

def stem_tokens(tokens):
    return [stemmer.stem(item) for item in tokens]

'''remove punctuation, lowercase, stem'''
def normalize(text):
    return stem_tokens(nltk.word_tokenize(text.lower().translate(remove_punctuation_map)))

vectorizer = TfidfVectorizer(tokenizer=normalize, stop_words='english')

def cosine_sim(text1, text2):
    tfidf = vectorizer.fit_transform([text1, text2])
    return ((tfidf * tfidf.T).A)[0,1]
```

```python
#comparaison des outputs de références et des outputs scrappées
title_score = []
text_score = []

for index, row in comparaison_df_small.iterrows():
    title1 = row["title"]
    title2 = row["scrapped_title"]
    title_score.append(cosine_sim(title1, title2))
    text1 = row["text"]
    text2 = row["scrapped_text"]
    text_score.append(cosine_sim(text1, text2))

print(title_score)
comparaison_df_small["title similarity"] = title_score
comparaison_df_small["text similarity"] = text_score
```
